## Non-linear filtering

**<a id="#">Task 3.2.1.</a>** Implement a median filter:

```python
def medianfilter(img_in, size):
    result = np.zeros(img_in.shape)

    k = size // 2
    h, w = img_in.shape

    for i in range(h):
        for j in range(w):

            values = []

            for di in range(-k, k + 1):
                for dj in range(-k, k + 1):

                    ni = min(max(i + di, 0), h - 1)
                    nj = min(max(j + dj, 0), w - 1)

                    values.append(img_in[ni, nj])

            result[i, j] = np.median(values)

    return result
```
